```python
n=int(input())

if n <= 1:
  print("Not valid")
elif n == 2:
  print("Prime")
else:
  is_prime = True
  for i in range(2, n):
    if n % i == 0:
      is_prime = False
      print("Not prime")
      break
  if is_prime:
    print("Prime")
```

```python
def count_vowels_consonants(text):
    vowels = 0
    consonants = 0
    lower_text = ""
    for ch in text:
        if 'A' <= ch <= 'Z':
            lower_text += chr(ord(ch) + 32)
        else:
            lower_text += ch
    for ch in lower_text:
        if 'a' <= ch <= 'z':
            if ch in ['a', 'e', 'i', 'o', 'u']:
                vowels += 1
            else:
                consonants += 1

    print("Vowels:", vowels)
    print("Consonants:", consonants)
user_input = input("Enter a string: ")
count_vowels_consonants(user_input)
```

```python
def remove_duplicates(lst):
    unique_list = []
    for item in lst:
        found = False
        for u in unique_list:
            if item == u:
                found = True
                break
        if not found:
            unique_list.append(item)
    return unique_list
my_list = list(map(int,input().split()))
print("Original list:", my_list)
print("List without duplicates:", remove_duplicates(my_list))

```

```python
def second_largest(lst):
    if len(lst) < 2:
        return None
    first = second = float('-inf')
    for num in lst:
        if num > first:
            second = first
            first = num
        elif num > second and num != first:
            second = num
    return second if second != float('-inf') else None
my_list = list(map(int,input().split()))
print("List:", my_list)
print("Second largest element:", second_largest(my_list))
```
